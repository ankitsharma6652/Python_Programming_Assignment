```python
string=input("Enter String:")
l=string.split()
u=[]
for i in l:
    if i in u:
        continue
    else:
        u.append(i)
print(" ".join(sorted(u)))

```

```python
string=input("Enter String:")
letters=0
digits=0
for i in string:
    if i.isnumeric():
        digits+=1
    elif i.isalpha():
        letters+=1

print(f"LETTERS {letters} \nDIGITS {digits}")
```

```python
password=input("Enter the Passwords seperated by comma").split(',')
import re 
p=[]
for i in password:
    if len(i)<6 or len(i)>12:
        continue
    elif not re.search("([A-Z a-z])+",i):
        continue
    elif not re.search("([0-9])+",i):
        continue
    elif not re.search("([$#@])+",i):
        continue
    else:
        p.append(i)
```
